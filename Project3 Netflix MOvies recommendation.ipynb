# Vectoriztion

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer(stop_words='english',binary=True,max_features=4000)   
```

```python
vector = cv.fit_transform(new_netflix['description']).toarray()
```

# Smimilarity Check

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    movie_index = new_netflix[new_netflix['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_netflix.iloc[i[0]]['title'])
```

```python
recommend('1920')
```

```python
recommend('3 Heroines')
```
